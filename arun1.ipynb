```python
plt.figure(figsize=(10,10))
plt.hist(x=df["Age"],color="red")

plt.title("Distribution of age",fontsize=20,color='blue')

plt.xlabel("Age",fontsize=15,color="red")
plt.ylabel("Frequency",fontsize=16,color='black')
plt.show()
```

```python
sns.boxplot(df["Age"])
plt.title("Distribution of Age")
plt.xlabel("Age")

plt.show()
```

```python
sns.boxplot(df['PassengerId'])
plt.title("Distribution of Age")
plt.xlabel("Age")

plt.show()
```

```python
df["Sex"].value_counts().plot(kind="barh",color='teal')
plt.title("Sex Distribution")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
df["Pclass"].value_counts().plot(kind="barh",color='teal')
plt.title("Pclass Distribution")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

```python
from matplotlib import cm
emb=df["Embarked"].value_counts()
keys=emb.keys().to_list()
counts=emb.to_list()
cs=cm.Set1([2,4,6,8])
plt.pie(x=counts,labels=keys,autopct='%1.1f%%',colors=cs)
plt.show()
```
